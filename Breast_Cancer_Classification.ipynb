Import The Dependencies

```python
from sklearn import datasets
```

Data Collection

```python
breast_cancer_dataset = datasets.load_breast_cancer()
```

```python
# Convert this into pandas DataFrame

data = pd.DataFrame(data = breast_cancer_dataset.data , columns = breast_cancer_dataset.feature_names)
```

Seprating The Features And Target

```python
X = data.drop('label' , axis =1)
y = data['label']
```

Splitting The Data Into Training Data And Testing Data

```python
from sklearn.model_selection import train_test_split
```

```python
X_train , X_test , y_train , y_test = train_test_split(X,y , test_size = 0.20 , random_state = 42)
```

```python
print(X_train.shape, X_test.shape , y_train.shape , y_test.shape)
```

Standardized The Data

```python
from sklearn.preprocessing import StandardScaler
sc= StandardScaler()

X_train_scaled = sc.fit_transform(X_train)
X_test_scaled   = sc.transform(X_test)
```

Importing Tesnsorflow Library

```python
import tensorflow as tf
from tensorflow import keras
tf.random.set_seed(3)
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense , Flatten
```

Setting Up Neural Network Layers

```python
model = Sequential()
model.add(Flatten(input_shape = (30,)))
model.add(Dense(20, activation = 'relu'))
model.add(Dense(2 , activation  = 'sigmoid'))
```

Compiling Neural Network

```python
model.compile(optimizer = 'adam' , loss= 'sparse_categorical_crossentropy' , metrics = ['accuracy'])
```

```python
history = model.fit(X_train_scaled ,y_train , validation_split = 0.1 , epochs = 10)
```

Visualizaing Loss And Accuracy

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history["val_accuracy"])
plt.title('Model_Accuracy')
plt.xlabel('epochs')
plt.ylabel("Accuracy")
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history["val_loss"])
plt.title('Model_Loss')
plt.xlabel('epochs')
plt.ylabel("Loss")
```

Accuracy Of Model On Test Data

```python
loss , accuracy = model.evaluate(X_test_scaled , y_test)
print(accuracy)
```
